```python
fig = plt.figure(1)
ax=fig.add_subplot(1,1,1)
ax.set_title('decision stumps in two variables')

def vertical(x,color):
    plt.plot([x,x], [0,1], c=color, label='x>'+str(x))
    plt.arrow(x,0.5,0.02,0, head_width=0.03, head_length=0.015, color=color)

def horizontal(y,color):
    plt.plot([0,1], [y,y], c=color, label='y>'+str(y))
    plt.arrow(0.5,y,0,0.02, head_width=0.015, head_length=0.04, color=color)

    
vertical(0.1,'k')
vertical(0.4, 'b')
horizontal(0.2, 'r')
horizontal(0.7, 'g')
plt.legend()
plt.xlim(-0.1,1.1)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
fig = plt.figure(1)
ax=fig.add_subplot(1,1,1)
ax.set_title('Beslutsstubbar i två variabler')

def vertical(x,color):
    plt.plot([x,x], [0,1], c=color, label='x>'+str(x))
    plt.arrow(x,0.5,0.02,0, head_width=0.03, head_length=0.015, color=color)

def horizontal(y,color):
    plt.plot([0,1], [y,y], c=color, label='y>'+str(y))
    plt.arrow(0.5,y,0,0.02, head_width=0.015, head_length=0.04, color=color)

    
vertical(0.1,'k')
vertical(0.4, 'b')
horizontal(0.2, 'r')
horizontal(0.7, 'g')
plt.legend()
plt.xlim(-0.1,1.1)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
fig = plt.figure(1)
ax=fig.add_subplot(1,1,1)
ax.set_title('decision trees in two variables, single tree shown')

def box(ax,x,y):
    pat = plt.Rectangle([x,y],0.1,0.1)
    ax.add_patch(pat)

for i in range(10):
    for j in range(10):
        if np.random.randn()<0.5:
            box(ax,i/10,j/10)

ax.set_aspect('equal')
plt.show()
```
